```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(n_samples =500 , centers = 4 , cluster_std = 0.9 , random_state= 0)

```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(6,6))
for cluster_id in np.unique(y_true):
  plt.scatter(
        X_scaled[y_true == cluster_id, 0 ] ,
        X_scaled[y_true == cluster_id, 1] ,
        label = f'True cluster {cluster_id}' ,
        alpha =0.6
               )
plt.legend()
plt.show()
```

```python
wcss = []
K_range  = range (1,11)
for k in K_range:
  km = KMeans(n_clusters = k , random_state = 0)
  km.fit(X_scaled)
  wcss.append(km.inertia_)

plt.figure(figsize = (6,6))
plt.plot(K_range, wcss , marker = 'o')
plt.xlabel('Number of Clusters k')
plt.ylabel('wcss (Intertia)')
plt.title('Elbow Method on Synthetic Blobs')
plt.show()
```
