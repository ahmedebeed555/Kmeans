```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample supermarket data: [Age, Spending]
X = np.array([
    [18, 150],  # Customer A
    [20, 200],  # Customer B
    [45, 800],  # Customer C
    [46, 850],  # Customer D
    [30, 400],  # Customer E
    [50, 1200], # Customer F
    [11, 34],
    [54, 455],
    [43,554],
    [65,775],
    [44, 2000],
    [44, 2100],
    [44,2200]

])

# Apply KMeans (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Plot clusters
plt.figure(figsize=(7,6))
for cluster_id in np.unique(y_kmeans):
    plt.scatter(
        X[y_kmeans == cluster_id, 0],
        X[y_kmeans == cluster_id, 1],
        label=f'Cluster {cluster_id}', alpha=0.7
    )

# Plot centroids
plt.scatter(centers[:,0], centers[:,1], c='red', s=200, marker='X', label='Centroids')

plt.xlabel("Age")
plt.ylabel("Monthly Spending ($)")
plt.title("Customer Segmentation with KMeans")
#plt.legend()
plt.show()
```
